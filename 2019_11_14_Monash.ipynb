```python
import requests
import PyPDF2

url = 'http://docs.autismresearchcentre.com/papers/2001_BCetal_AQ.pdf'

r = requests.get(url)

with open('./metadata.pdf', 'xb') as f:
    f.write(r.content)
```

```python
# you can find find the pdf file with complete code in below
with open('./metadata.pdf', 'rb') as pdfFileObj:
# pdf reader object
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
# number of pages in pdf
    print(pdfReader.numPages)
# a page object
    for page in range(pdfReader.numPages):
        print('\n\n****',page,'****\n\n')
        pageObj = pdfReader.getPage(page)
# extracting text from page.
# this will print the text you can also save that into String
        print(pageObj.extractText())
```

```python
text = pdfReader.getPage(1).extractText()
```

```python
text.find('social skill(items 1,11')
```

```python
text.find('. Each of the items listed above scores 1 point')
```

```python
subscores = text[2041:text.find('. Each of the items listed above scores 1 point')]
```

```python
subscores_dict = {}

for line in subscores.replace(' ','_').split(';'):
    k,v = line.split('(items_')

    if k.startswith('_'):
        k = k[1:]
    if v.endswith(')'):
        v = v[:-1]
    subscores_dict[k] = [int(i) for i in v.split(',')]
```

```python
def score_agree(x):
    return [1 if i in ['Slightly agree','Definitely agree'] 
            else 0 for i in x ]
```

```python
def score_disagree(x):
    return [1 if i in ['Slightly disagree','Definitely disagree'] 
            else 0 for i in x ]
```
